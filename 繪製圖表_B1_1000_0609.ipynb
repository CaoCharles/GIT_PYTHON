```python
# 匯入模組
import numpy as np
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\SimSun.ttc", size=14) 
plt.style.use('ggplot')
%matplotlib inline 
##離線
import plotly
from plotly.graph_objs import Scatter, Layout
plotly.offline.init_notebook_mode(connected = True)
```

## m1 = 200 m2 = 800

```python
# 繪圖
data_MC = pd.read_csv('./DATA/MC_DEP_CSN_1000_B1_0609.csv').replace(0, 'NA')
data_MIS = pd.read_csv('./DATA/MIS_DEP_CSN_1000_B1_0609.csv').replace(0, 'NA')
data_GIS = pd.read_csv('./DATA/GIS_DEP_CSN_1000_B1_0609.csv').replace(0, 'NA')

X = np.arange(10, 1210, 10)
# Create a trace
trace1 = go.Scatter(x = X, 
                    y = data_MC['估計值'],
                    name = 'MC',
                    connectgaps = True,
                    line = dict(
                                color = '#Fb6f6f',
                                width = 2,
                    )
                   )
                    
trace2 = go.Scatter(
                    x = X,
                    y =data_MIS['估計值'],
                    name = 'MIS',
                    connectgaps = True,
                    line = dict(
                                color = '#4d0b0b', 
                                width = 1.1,
                                dash = 'dashdot'
                    )
                   )
trace3 = go.Scatter(
                    x = X,
                    y = data_GIS['估計值'],
                    name = 'GIS',
                    connectgaps = True,
                    line = dict(
                                color = '#1088b2',
                                width = 2,
                                dash = 'dash')
)
datpy = [trace1,trace2, trace3]

layout = dict(plot_bgcolor = '#F1F1F1',
              xaxis = dict(title = 'Loss Level', 
                           showgrid = False, 
                           zeroline = False, 
                           showline = True, 
                           mirror=True,
                           tickvals = [0,200,400,600,800,1000,1200],
                           ticks = 'outside'),
              yaxis = dict(title = 'Tail Probablity ',
                           type = 'log',
                           exponentformat='e',
                           showgrid = False,
                           zeroline = False, 
                           showline = True, 
                           tickangle = 270,
                           mirror=True,
                           ticks = 'outside',
                           tickvals = [10**(-8),10**(-6),10**(-4),10**(-2) , 1],
                           ticktext = ['1e-8','1e-6','1e-4','1e-2','1',],
                           showexponent = "all", 
                          ),
              autosize=False,
              margin=go.layout.Margin(l=70,
                                      r=50,
                                      b=50,
                                      t=30,
                                      pad=4
                                    ),
              legend = dict(
                            orientation = "h", x = 0.01, y = 0.1,
                            font = dict(size=12,
                                        color='#000'
                                      ),
                           ),
              
             )
fig = go.Figure(data=datpy, layout=layout)
plotly.offline.iplot(fig)
```
